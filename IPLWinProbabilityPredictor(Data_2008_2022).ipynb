```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('IPL_Matches_2008_2022.csv')
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
match.head()
```

```python
delivery
```

```python
total_score_df = delivery.groupby(['ID', 'innings']).sum()['total_run'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['innings']==1]
total_score_df
```

```python
match_df = match.merge(total_score_df[['ID','total_run']])
```

```python
match_df
```

```python
match_df['Team1'].unique()
```

```python
teams = ['Rajasthan Royals', 
         'Royal Challengers Bangalore',
         'Sunrisers Hyderabad',
         'Delhi Capitals', 
         'Chennai Super Kings',
         'Gujarat Titans',
         'Lucknow Super Giants', 
         'Kolkata Knight Riders',
         'Punjab Kings', 
         'Mumbai Indians'
        ]
              
       
```

```python
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrises Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrises Hyderabad')



```

```python
match_df = match_df[match_df['Team1'].isin(teams)]
match_df = match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df = match_df[['ID','City', 'WinningTeam', 'total_run', 'Team1','Team2']]
```

```python
match_df
```

```python
delivery_df = match_df.merge(delivery,on='ID')
```

```python
delivery_df = delivery_df[delivery_df['innings']==2]
```

```python
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby('ID').cumsum()['total_run_y']
```

```python
delivery_df.head(220)
```

```python
delivery_df['runs_left'] = delivery_df['total_run_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] =  120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
delivery_df['player_out'] = delivery_df['player_out'].fillna("0")
```

```python
delivery_df['player_out'] = delivery_df['player_out'].apply(lambda x:x if x == "0" else "1")
```

```python
delivery_df['player_out'] = delivery_df['player_out'].astype('int')
delivery_df
```

```python
wickets = delivery_df.groupby('ID').cumsum()['player_out']
delivery_df['wickets'] = 10 - wickets
```

```python
delivery_df
```

```python
delivery_df.head()
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
def BowlingTeam(row):
    return row['Team2'] if row['BattingTeam'] == row['Team1'] else row['Team1']
```

```python
delivery_df['Team2'] = delivery_df.apply(BowlingTeam,axis=1)
```

```python
delivery_df['BowlingTeam'] = delivery_df['Team2']
```

```python
final_df= delivery_df[['BattingTeam','BowlingTeam','City','runs_left','balls_left','wickets','total_run_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
 #model building
final_df.dropna(inplace=True)    
```

```python
final_df = final_df[final_df['balls_left']!=0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','City'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)

```
